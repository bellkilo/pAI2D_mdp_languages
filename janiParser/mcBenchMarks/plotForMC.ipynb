```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("brp_test_results.csv")

df_n = df.sort_values(by="n")

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(df_n["n"], df_n["time_SD"], label="time_SD", marker='o')
plt.plot(df_n["n"], df_n["time_RLGL"], label="time_RLGL", marker='x')
plt.xlabel("n")
plt.ylabel("Time (s)")
plt.title("Execution Time vs n")
plt.legend()
plt.grid(True)

df_mn = df.sort_values(by="m/n")

plt.subplot(1, 2, 2)
plt.plot(df_mn["m/n"], df_mn["time_SD"], label="time_SD", marker='o')
plt.plot(df_mn["m/n"], df_mn["time_RLGL"], label="time_RLGL", marker='x')
plt.xlabel("m/n")
plt.ylabel("Time (s)")
plt.title("Execution Time vs m/n")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.scatter(df["n"], df["m/n"], df["time_SD"], c=df["time_SD"], cmap="viridis")
ax1.set_xlabel("n")
ax1.set_ylabel("m/n")
ax1.set_zlabel("time_SD")
ax1.set_title("3D: time_SD vs n and m/n")

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.scatter(df["n"], df["m/n"], df["time_RLGL"], c=df["time_RLGL"], cmap="plasma")
ax2.set_xlabel("n")
ax2.set_ylabel("m/n")
ax2.set_zlabel("time_RLGL")
ax2.set_title("3D: time_RLGL vs n and m/n")

plt.tight_layout()
plt.show()
```
